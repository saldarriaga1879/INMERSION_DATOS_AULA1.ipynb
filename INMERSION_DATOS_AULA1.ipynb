# Empezando a Cambiar de string a numerico con **expand=True**
# Ademas agrega nuevos Atributos

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['precio'] = valor[0]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
# Elilminar los puntos en el atributo PRECIO
inmuebles['precio'] = inmuebles['precio'].str.replace('.','',regex=True)
```

```python
# El 1er corchete representa la lista
# El 2do corchete (amarillo) hace q la computadora traiga la info como un DataFrame
inmuebles[['precio','Barrio']]
```

```python
inmuebles.info()
```

```python
#cambiar precio de tipo object a float
#creamos un nuevo Atributo
inmuebles['precio_millon'] = inmuebles.precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.head()
```

```python
round(inmuebles.describe(),2)
```

```python
# Otra manera de eliminar decimales, con funciones de Libreria Pandas
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
inmuebles.describe()
```
